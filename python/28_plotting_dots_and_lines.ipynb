```python
import matplotlib.pyplot as plt

plt.plot(3, 4, 'r^')
plt.plot(2, 4.2, 'bo')
```

```python
import matplotlib.pyplot as plt

plt.plot(3, 4, 'r^')
plt.show() # End plotting and start fresh
plt.plot(2, 4.2, 'bo')
plt.show()
```

```python
plt.plot(2, 4.2, 'bo', label='blue dot')
plt.plot(3, 4, 'r^', label='red triangle')
plt.legend() # Will show labels defined above in code order
plt.show()
```

```python
plt.plot(2, 4.2, 'bo')
plt.plot(3, 4, 'r^')
plt.legend(['blue dot', 'red triangle'])
plt.show()
```

```python
# draw a line
x = [0, 1, 3]
y = [0, 2, -1]

plt.plot(x, y, 'ms:') # instead of : we can use --
plt.show()
```

```python
import numpy as np

x = np.linspace(0, 3*np.pi, 101)
y = np.sin(x)

plt.plot(x, y, 'm:') # instead of : we can use --
plt.show()
```
